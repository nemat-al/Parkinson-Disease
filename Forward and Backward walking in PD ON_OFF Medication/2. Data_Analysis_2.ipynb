```python
pd_on_noraml_df_columns=pd_on_noraml_df.drop(['id','Full_Name','Reference'], axis=1)
```

```python
fa.fit(pd_on_noraml_df_columns)
# Check Eigenvalues
ev, v = fa.get_eigenvalues()
sorted(ev)
```

```python
plt.scatter(range(1,pd_on_noraml_df_columns.shape[1]+1),ev)
plt.plot(range(1,pd_on_noraml_df_columns.shape[1]+1),ev)
plt.title('Forward - ON - Scree plot')
plt.xlabel('Favtors')
plt.ylabel('Eigenvalue')
plt.grid()
plt.show()
```

```python
pd.DataFrame(fa.loadings_, columns=['Factor1','Factor2','Factor3'], index=[pd_on_noraml_df_columns.columns])
```
